```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

6.1.2

```python
import sympy as symengine
vars = symengine.symbols('x y') # Define x and y variables
f = symengine.sympify(['(x-x**3)', '-y']) # Define function
J = symengine.zeros(len(f),len(vars)) # Initialise Jacobian matrix
# Fill Jacobian matrix with entries
for i, fi in enumerate(f):
  for j, s in enumerate(vars):
    J[i,j] = symengine.diff(fi, s)
print(J)
```

evaluamosen en los puntos (0,0)

```python
b= np.array([[1,0],[0,-1]])
c= np.array([0,0])
d = np.linalg.det(b)
t=b.trace()
l1=(t+np.sqrt(t**2-4*d))/2
l2=(t-np.sqrt(t**2-4*d))/2
print(l1,l2)
```

```python
def fun (t,xy,a,b):
  x,y=xy
  return [x-x**3,-y]
  
a,b=-2,2

ti=0; tf=10; dt=1000
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-5,5,1):
  for y0 in np.arange(-5,5,1):
    sol=solve_ivp(fun,tspam,[x0,y0],args=(a,b),t_eval=t)
    plt.plot(sol.y[0], sol.y[1])

plt.axis([-5, 5, -5, 5])
plt.grid()
```

```python
# Funcion para graficar el retrato de fase de cualquier sistema

def plot_phase_portrait(fun):
  t_span = [0,15]
  plt.figure(figsize=(16,9))

  for x0 in np.arange(0,3,0.3):
    for y0 in np.arange(0,3,0.3): 
        y_0 = [x0,y0]
        t = np.arange(0,15,0.01)
        sol = solve_ivp(fun, t_span = t_span, y0 = y_0,t_eval=t)

        plt.plot(sol.y[0],sol.y[1])
        plt.title("Phase portrait",fontsize=20)
        plt.xlabel("$x$",fontsize=20)
        plt.ylabel("$y$",fontsize=20)
  plt.show()
```
