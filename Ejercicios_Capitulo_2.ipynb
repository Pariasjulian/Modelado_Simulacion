```python
# Importamos las librerías

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definición de funciones

# Función del método de euler
def euler_method(f, h, x0, ti, tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    x[i+1] = x[i] + h*f(x[i])

  t = np.linspace(ti,tf,N)
  
  return t,x

# Función para graficar el phase portrait y el vector field
def plot_phase_portrait(f,step,ti,tf,x0_i,x0_f,n):
  
  plt.figure(figsize=(20,9))

  plt.subplot(121)
  for i in np.linspace(x0_i,x0_f,n):
    x0 = i
    t,x = euler_method(f,step,x0,ti,tf)
    plt.plot(t,x,label=f'$x_0 = {i}$')
    plt.xlabel("$t$",fontsize=15)
    plt.ylabel("$x(t)$",fontsize=15)
    plt.title("Phase Portrait",fontsize=15)
    # plt.legend()
  plt.subplot(122)
  x_plot=np.linspace(x0_i,x0_f,100)
  plt.plot(x_plot,f(x_plot))
  plt.xlabel("$x$",fontsize=15)
  plt.ylabel("$\dot x$",fontsize=15)
  plt.title("Vector Field",fontsize=15)
  plt.show()
```

### 2.1.1: Find all the fixed points of the flow.

Para que un punto sea fijo, el flujo ($sin(x)$) tiene que ser 0.

$\dot x = 0 ⇔ sin (x) = 0 ⇒ x∗=nπ$   $ ∀n ∈ N$

Existen infinitos puntos fijos.


```python
def func21(x):
  return np.sin(x)

h = 1e-3
ti = 0
tf = 30
x0_i = -2*np.pi
x0_f = 2*np.pi
n = 20

plot_phase_portrait(func21,h,ti,tf,x0_i,x0_f,10)
```

```python
def func11(x):
  return -x

step = 1 # delta t
x0 = 1 # condición inicial x(0) = 1
ti = 0 # tiempo inicial
tf = 1 # tiempo final x(1)

real_solution = np.exp(-1)
deltas = np.array([])
euler_errors = np.array([])

for n in range(5):
  step = 10**(-n)
  deltas = np.append(deltas,step)
  t,x = euler_method(func11,step,x0,ti,tf)
  euler_errors = np.append(euler_errors,abs(x[-1] - real_solution))

plt.figure(figsize=(16,9))
plt.subplot(121)
plt.plot(deltas,euler_errors)
plt.title("$E$ as a function of $\Delta t$",fontsize=15)
plt.xlabel("$\Delta t$",fontsize=15)
plt.ylabel("E",fontsize=15)

plt.subplot(122)
plt.plot(np.log(deltas),np.log(euler_errors))
plt.title("$\ln (E)$ as a function of $\ln (\Delta t)$",fontsize=15)
plt.xlabel("$\ln \Delta t$",fontsize=15)
plt.ylabel("$\ln (E)$",fontsize=15)


plt.show()
```

Como se puede ver, a medida que se aumenta el tamaño del paso, el error crece, pues la aproximación se hace cada vez más mala. Además, se puede ver que el error crece (o decrece) de forma lineal conforme se cambia el tamaño del paso. En pocas palabras, podemos concluir que el error y el tamaño del paso son directamente proporcionales.

$E \propto \Delta t$
